```python
import numpy as np
import pylab as pl
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
from sklearn.utils import shuffle
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

```

```python
train1 = pd.read_csv(r"C:\Users\Saranga\Desktop\FYP\My DNN\training_set_my.csv",header=None)

```

```python
train1.drop(labels=[0],axis=0,inplace=True)
train1
```

```python
test1 = pd.read_csv(r"C:\Users\Saranga\Desktop\FYP\My DNN\testing_set_my.csv",header=None)
```

```python
test1.drop(labels=[0],axis=0,inplace=True)
test1
```

```python
train = shuffle(train1)
test = shuffle(test1)
train
```

```python
X_train = train.iloc[:,2:53].values
X_train 
```

```python
X_test = test.iloc[:,2:53].values
X_test 
```

```python
Y_train_label = train.iloc[:,53].values.astype(int)
Y_train_label
```

```python
Y_test_label = test.iloc[:,53].values.astype(int)
#Y_test_label = Y_test_label.astype(np.int)
Y_test_label
```

```python
X_train_scaled = (X_train - X_train.min()) / (X_train.max() - X_train.min())
X_test_scaled = (X_test - X_test.min()) / (X_test.max() - X_test.min())
X_train_scaled
```

```python
params_grid = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
```

```python
svm_model = GridSearchCV(SVC(), params_grid, cv=5)
svm_model.fit(X_train_scaled, Y_train_label)
```

```python
print('Best score for training data:', svm_model.best_score_,"\n") 
```

```python
print('Best C:',svm_model.best_estimator_.C,"\n") 
print('Best Kernel:',svm_model.best_estimator_.kernel,"\n")
print('Best Gamma:',svm_model.best_estimator_.gamma,"\n")

```

```python
final_model = svm_model.best_estimator_
Y_pred_label = final_model.predict(X_test_scaled)
Y_pred_label
```

```python
Y_test_label
```

```python
print(classification_report(Y_test_label,Y_pred_label))

```

```python
print("Training set score for SVM: %f" % final_model.score(X_train_scaled , Y_train_label))
print("Testing  set score for SVM: %f" % final_model.score(X_test_scaled  , Y_test_label ))

svm_model.score
```

```python
print(confusion_matrix(Y_test_label,Y_pred_label))
```
